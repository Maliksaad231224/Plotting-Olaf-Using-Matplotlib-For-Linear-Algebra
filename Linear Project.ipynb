```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.image as mpimg
import numpy as np
from scipy.interpolate import interp1d

# Load the background image
bg_img = mpimg.imread('olaf bg.png')  # Or .jpg




fig, ax = plt.subplots(figsize=(10, 10))

ax.imshow(bg_img, extent=[-15, 15, -15, 15], zorder=0)

# Define body coordinates
x_coord = [-0.8, -0.6,   0,    1,    2,    2,    2,  1.7,    1,  0.3,   -2,   -3,   -4,   -4, -4.5  ]
y_coord = [-0.8, -1.2, -1.3, -1.4,   -1,    0,    7,    9,   12, 12.4,   15, 14.5,   13,  8.5,  7.5 ]

plt.fill(x_coord, y_coord, color='#f0f8ff', edgecolor='black', linewidth=1)
plt.plot(x_coord, y_coord, color='black', linewidth=1)

plt.axis('equal')  # Keep proportions correct
plt.axhline(0, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')
plt.xticks(range(-15, 16, 1))
plt.yticks(range(-15, 16, 1))
plt.xlim(-15, 15)
plt.ylim(-15, 15)
plt.grid(True)


# Add coordinate labels along the central axes
for i in range(-15, 16):
    if i != 0:  # Skip 0 to avoid overlap
        # X-axis labels (y=0)
        ax.text(i, 0.5, str(i), ha='center', va='bottom', color='black', fontsize=8)
        # Y-axis labels (x=0)
        ax.text(0.5, i, str(i), ha='left', va='center', color='black', fontsize=8)
plt.gca().set_aspect('equal', adjustable='box')


# eyes


eye1 = patches.Circle((-1.5,11.5), radius=0.6, edgecolor='black', facecolor='white', linewidth =3)
eye2 = patches.Circle((-3, 10.5), radius=0-.6, edgecolor='black', facecolor='white', linewidth =3)   
ax.add_patch(eye1)
ax.add_patch(eye2)

pupil1 = patches.Circle((-1.5, 11.5), radius=0.2, edgecolor='black', facecolor='black') 
pupil2 = patches.Circle((-3, 10.5), radius=0.2, edgecolor='black', facecolor='black')  
ax.add_patch(pupil1)
ax.add_patch(pupil2)

#nose

x_nose = [-1.5, -3.0, -3.5, -3.8, -3.5, -3.0, -2.0, -1.2]  
y_nose = [10.5, 10.0, 9.0, 7.5, 7.0, 7.5, 8.0, 9.0] 

t = np.linspace(0, 1, len(x_nose))
t_smooth = np.linspace(0, 1, 200) 
cs_x = interp1d(t, x_nose, kind='cubic')
cs_y = interp1d(t, y_nose, kind='cubic')
x_smooth = cs_x(t_smooth)
y_smooth = cs_y(t_smooth)

plt.fill(x_smooth, y_smooth, color='#FD6A02', edgecolor='black', linewidth=2, zorder=3)
plt.plot(x_smooth, y_smooth, color='black', linewidth=2, zorder=4)

shadow_x = np.linspace(min(x_smooth), max(x_smooth), 50)
shadow_y = np.linspace(min(y_smooth), min(y_smooth) - 0.3, 50)
plt.fill_between(shadow_x, shadow_y, min(y_smooth), color='black', alpha=0.1, zorder=2)


#smile

x_smile = [-2.7, -1,0,0.7,1,1,0.7,0,-1,-2, -3.1    ]
y_smile = [ 4.8,  1,1,2  ,3,8,9  ,7,6 ,5.2,5.2]


plt.fill(x_smile, y_smile, color='#3e5a69', edgecolor='black', linewidth=1)
plt.plot(x_smile, y_smile, color='black', linewidth=1)

#teeth 
x_teeth = [-2.4 ,-2, -0.2,-0.1,-0.4]
y_teeth = [5    ,4, 5.3 , 6,6.5]


plt.fill(x_teeth, y_teeth, color='#f0f8ff', edgecolor='black', linewidth=1)
plt.plot(x_teeth, y_teeth, color='black', linewidth=1)



#chest
x_chest = [-0.8,-1.2,-1.5,-1.6,-1.6,0 , 1.8  ,2.5 ,3,3.2    , 3.3,3    ,2.5,2,1,0,-0.8]
y_chest = [-0.8,-1.2,-2  ,-2  ,-2.5,-5.5,-8.4,-6.5,-5.5,-4.5,-3.5,-2.7,-1.7,-1,-1.5,-1.4,-0.8]

plt.fill(x_chest, y_chest, color='#f0f8ff', edgecolor='black', linewidth=2)
plt.plot(x_chest, y_chest, color='black', linewidth=1)


#body 
x_body = [-1.6,-2  ,-3  ,-4,-5  ,-5.2,-5.5,-5.7,-5.7,-5.5,-5.0,-4,-2.5,1,1.5,  1.9   ,2.2  ,2.4  ,2.2 ,2,1.8]
y_body = [-2.5,-2.5,-2.7,-3,-4.3,-4.6,-5.5,-6.5,-7.5,-9.5,-10.9,-13,-15,-15,-14,-12.8,-11.5,-10.5,-9.7,-9,-8.5]

plt.fill(x_body, y_body, color='#f0f8ff', edgecolor='black', linewidth=2)
plt.plot(x_body, y_body, color='black', linewidth=1)

#feet

x_foot = [-5.7, -6.5, -7.5, -7.8, -7.9, -7.7,-7.5,-7.1,-7   ,-6.8 ,-6.2 ,-5.2 ,-5.3,-5.5]
y_foot = [-7.2, -6.6, -6.7, -7.5, -8  , -9  ,-10 ,-11 ,-11.2,-11.5,-11.5,-10.7,-10,-9]
plt.fill(x_foot, y_foot, color='#f0f8ff', edgecolor='black', linewidth=2)
plt.plot(x_foot, y_foot, color='black', linewidth=2)

#buttons
button1= patches.Circle((0.5, -3.5), radius=1, edgecolor='black', facecolor='black') 
button2 = patches.Circle((-1.3, -8.5), radius=1, edgecolor='black', facecolor='black')  
button3 = patches.Circle((-3, -10.5), radius=1, edgecolor='black', facecolor='black')  
ax.add_patch(button1)
ax.add_patch(button2)
ax.add_patch(button3)
plt.grid(True)
plt.show()
```

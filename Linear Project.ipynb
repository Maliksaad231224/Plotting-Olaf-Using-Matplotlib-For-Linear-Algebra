```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.image as mpimg

# Load the background image
bg_img = mpimg.imread('olaf bg.png')  # Or .jpg


# Define body coordinates
x_coord = [2, 4, 4, 0, -3, -3, -4, -4, -3, 0, 2]
y_coord = [-2, -1, 6, 10, 8, 6, 5, 4, 2, -1, -2]

fig, ax = plt.subplots(figsize=(10, 10))

ax.imshow(bg_img, extent=[-15, 15, -15, 15], zorder=0)


plt.fill(x_coord, y_coord, color='#f0f8ff', edgecolor='black', linewidth=2)
plt.plot(x_coord, y_coord, color='black', linewidth=2)



eye1 = patches.Circle((-1.5, 6.5), radius=1, edgecolor='black', facecolor='white')
eye2 = patches.Circle((0, 7), radius=1, edgecolor='black', facecolor='white')   
ax.add_patch(eye1)
ax.add_patch(eye2)

pupil1 = patches.Circle((-1.5, 6.5), radius=0.3, edgecolor='black', facecolor='black') 
pupil2 = patches.Circle((0, 7), radius=0.3, edgecolor='black', facecolor='black')  
ax.add_patch(pupil1)
ax.add_patch(pupil2)



plt.axhline(0, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')
plt.xticks(range(-15, 16, 1))
plt.yticks(range(-15, 16, 1))
plt.xlim(-15, 15)
plt.ylim(-15, 15)
plt.grid(True)

plt.gca().set_aspect('equal', adjustable='box')

x_nose = [0, -1, -2, -3]
y_nose = [5,  6,  6, 3]


plt.fill(x_nose, y_nose, color='#FD6A02', edgecolor='black', linewidth=2)
plt.plot(x_nose, y_nose, color='black', linewidth=2)

x_smile = [-3, 0, 2, 3, 2, 1,0,-1]
y_smile = [ 2, 2, 4, 6, 1,0 ,0, 0]


plt.fill(x_smile, y_smile, color='#3e5a69', edgecolor='black', linewidth=2)
plt.plot(x_smile, y_smile, color='black', linewidth=2)


x_teeth = [-2,-1,1,1,0]
y_teeth = [2,1.2,2,2.8,2]


plt.fill(x_teeth, y_teeth, color='#f0f8ff', edgecolor='black', linewidth=2)
plt.plot(x_teeth, y_teeth, color='black', linewidth=2)



x_body = [0,0  , 3 ,5 ,6,5, 4 ,2 ]
y_body = [-1,-4,-6,-6,-5,-2,-1,-2 ]

plt.fill(x_body, y_body, color='#f0f8ff', edgecolor='black', linewidth=2)
plt.plot(x_body, y_body, color='black', linewidth=2)


x_lower_body = [5, 7,  7, 6  , 4,4,2]
y_lower_body = [-6,-7,-9, -11,-11,-10,-9]

plt.fill(x_lower_body, y_lower_body, color='#f0f8ff', edgecolor='black', linewidth=2)
plt.plot(x_lower_body, y_lower_body, color='black', linewidth=2)

hair_color = '#4B2E2B' 
plt.plot([0, 0], [10, 13], color=hair_color, linewidth=2)
plt.plot([0, -1], [10, 13], color=hair_color, linewidth=2)
plt.plot([0, 1], [10, 13], color=hair_color, linewidth=2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.image as mpimg
import numpy as np

# Load the background image
bg_img = mpimg.imread('3o9q7j2of9y81 (1).png')

# Set up the plot
fig, ax = plt.subplots(figsize=(10, 10))

# Display the image with the coordinate extent
ax.imshow(bg_img, extent=[-15, 15, -15, 15], zorder=0)

# Add x and y axis lines at 0
ax.axhline(0, color='black', linewidth=1)
ax.axvline(0, color='black', linewidth=1)

# Set aspect ratio
ax.set_aspect('equal')

# Set ticks every 1 unit
ticks = np.arange(-15, 16, 1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)

# Add grid
ax.grid(True, which='both', linestyle='--', linewidth=0.5,color='black')

# Set limits
ax.set_xlim(-15, 15)
ax.set_ylim(-15, 15)

# Show the plot
plt.show()
```
